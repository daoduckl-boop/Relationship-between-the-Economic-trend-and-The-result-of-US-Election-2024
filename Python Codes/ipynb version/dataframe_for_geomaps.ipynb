```python
df_newdelta = pd.read_csv('df_newdelta.csv') # upload file
df_newdelta=df_newdelta.set_index('GeoName')
df_newdelta
```

```python
df_election_24= pd.read_csv('election_percentages_24.csv')
df_election_24=df_election_24.rename(columns={'Unnamed: 0':'GeoName'}) # rename column
df_election_24=df_election_24.set_index('GeoName') #set index
df_election_24
```

```python
df_merge = pd.merge(df_newdelta,df_election_24,on='GeoName') # merge two data
df_merge
```

```python
df_merge['Result']=df_merge.apply(lambda x:'REPUBLICAN' if x['REPUBLICAN']>x['DEMOCRAT'] else 'DEMOCRAT',axis=1) # find the win party
```
